```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as bs
from pprint import pprint # pour afficher clairement la liste
```

```python
#lien du site a scrape
url = "https://books.toscrape.com/"
response = rq.get(url)
soup = bs(response.text, "html.parser")
#print(soup.prettify()) #prettify est utilise pour le formatage du code html
```

## fonction pour parcourir recurcivement l'arbre DOM

```python
def traverse_dom(element, level=0):
    #afficher l'element actuel
    if element.name:
        print(f"{' '* level}<{element.name}>")
    #si l'element a des enfants le parcourir egalement
    if hasattr(element, 'children'):
        for child in element.children:
            traverse_dom(child, level + 1)
# commencer le parcours depuis la racine de l'arbre DOM
traverse_dom(soup)
```
