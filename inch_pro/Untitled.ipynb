```python
#importations des donnees
import numpy as np 
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests as rq
import csv
```

```python
#recuperation du code html avec requests et beautifulSoup
url = "https://www.planete-energies.com/fr/media/chiffres/production-mondiale-petrole"
r = rq.get(url)
soup = bs(r.text, "lxml")
#soup = bs(r.content, "html.parser")
soup
```

```python
# extractions des entetes ou titres du table
headers = soup.find_all("th")
print(headers)
titles = []
for i in headers:
    title = i.text
    titles.append(title)
titles
```

```python
# extration des infos du table
rows = soup.find_all("tr")

row_list = []
for i in rows[1:]:
    data = i.find_all("td")
    row = [td.text for td in data]
    row_list.append(row)
row_list    
```

```python
#creation d'une nouvelle list a partir des anciennes listes
entire = [titles] + row_list
entire
```

```python
# creation du DataFrame  avec pandas
df = pd.DataFrame(columns = entire)
df
```

```python
# extraction sous forme de fichier csv avec pandas
df.to_csv("data.csv",index = False)
```
